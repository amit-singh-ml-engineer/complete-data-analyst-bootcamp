```python
## We can fill the average mean of the age
data=[
    {'Name':'Amit','Age':20,'City':'Bangalore'},
    {'Name':'Ashish','Age':21,'City':'New York'},
    {'Name':'Jack','City':'Bijing'},
    {'Name':'John','Age':23,'City':'Florida'}
]
df=pd.DataFrame(data)
df=df.fillna(df['Age'].mean())
print(df)
```

```python
## We can change the data type
import pandas as pd
data=[
    {'Name':'Amit','Age':20,'City':'Bangalore'},
    {'Name':'Ashish','Age':21,'City':'New York'},
    {'Name':'Jack','City':'Bijing'},
    {'Name':'John','Age':23,'City':'Florida'}
]
df=pd.DataFrame(data)
df=df['Age'].fillna(df['Age'].mean()).astype(int)
print(df)
```

```python
## We can multiply the integer part
import pandas as pd
data=[
    {'Name':'Amit','Age':20,'City':'Bangalore'},
    {'Name':'Ashish','Age':21,'City':'New York'},
    {'Name':'Jack','City':'Bijing'},
    {'Name':'John','Age':23,'City':'Florida'}
]
df=pd.DataFrame(data)
df['Age']=df['Age'].fillna(df['Age'].mean()).astype(int)
df['Square_of_age']=df['Age'].apply(lambda x:x**2)
print(df)
```

```python
## We can merge and joine two DataFrame(inner)
df1=pd.DataFrame({'key':['A','B','C'],
                  'value1':[1,2,3]
                  })
df2=pd.DataFrame({'key':['A','B','D'],
                  'value2':[4,5,6]
                  })
pd.merge(df1,df2,on="key",how="inner")
```

```python
## We can merge and joine two DataFrame(outer)
df1=pd.DataFrame({'key':['A','B','C'],
                  'value1':[1,2,3]
                  })
df2=pd.DataFrame({'key':['A','B','D'],
                  'value2':[4,5,6]
                  })
pd.merge(df1,df2,on="key",how="outer")
```

```python
## We can merge and joine two DataFrame(right)
df1=pd.DataFrame({'key':['A','B','C'],
                  'value1':[1,2,3]
                  })
df2=pd.DataFrame({'key':['A','B','D'],
                  'value2':[4,5,6]
                  })
pd.merge(df1,df2,on="key",how="right")
```

```python
## We can merge and joine two DataFrame(left)
df1=pd.DataFrame({'key':['A','B','C'],
                  'value1':[1,2,3]
                  })
df2=pd.DataFrame({'key':['A','B','D'],
                  'value2':[4,5,6]
                  })
pd.merge(df1,df2,on="key",how="left")
```
