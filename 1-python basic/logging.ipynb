```python
import logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

```python
import logging
#create a logger for module 1
logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)
## create a logger for module 2
logger2=logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

## configure logging setting

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%M-%d %H:%M:%S'
)

## log message with diffrent module
logger1.debug("This is a debug message for module 1")
logger2.warning("This is warning message for module 2")
logger2.error("This is error message")


```
