HANDALING IMBALANCED DATASET

1 Up Sampling

2 Down Sampling

```python
import numpy as np
import pandas as pd

## Set the random seed for reproducibility
np.random.seed(123)


## Create a DataFrame with two class
n_Samples=1000
class_0_ratio=0.9
n_class_0=int(n_Samples*class_0_ratio)
n_class_1=n_Samples-n_class_0

```

```python
n_class_0,n_class_1
```

```python
## Create my Dataframe with imbalance dataset
class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
##Up sampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True, ##Sample with replacement
                               n_samples=len(df_majority),
                               random_state=42
                               )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df=df_upsample=pd.concat([df_majority,df_minority_upsampled])
```

```python
df['target'].value_counts()
```

```python
## Down Sampling
import numpy as np
import pandas as pd

## Set the random seed for reproducibility
np.random.seed(123)


## Create a DataFrame with two class
n_Samples=1000
class_0_ratio=0.9
n_class_0=int(n_Samples*class_0_ratio)
n_class_1=n_Samples-n_class_0


## Create my Dataframe with imbalance dataset
class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})



df=pd.concat([class_0,class_1]).reset_index(drop=True)

##Check the class distribution

print(df['target'].value_counts())
```

```python
##Down sampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_majority_downsampled=resample(df_majority,replace=False,
                               n_samples=len(df_minority),
                               random_state=42
                               )
```

```python
df_majority_downsampled.shape
```

```python
df_majority_downsampled.head()
```

```python
df_down=pd.concat([df_minority,df_majority_downsampled])
```

```python
df_down['target'].value_counts()
```
